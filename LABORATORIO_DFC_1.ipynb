#1 ETL ( Extracción, Transformación y carga)

```python
from google.colab import drive
import pandas as pd
import chardet
ruta_archivo = '/content/drive/MyDrive/transactions.csv'
with open(ruta_archivo, 'rb') as f:
    result = chardet.detect(f.read())
    print(result)
try:
  df= pd.read_csv(ruta_archivo, encoding=result['encoding'], sep=',', decimal='.')
  print("Extraccion de datos exitoosa")
except FileExistsError:
  print("Error en la extraccion de datos")
```

```python
df.info()
```

```python
import pandas as pd

#Separación de fehca y hora  la columna transation_time

df['transaction_time'] = pd.to_datetime(df['transaction_time'])

df['date'] = df['transaction_time'].dt.date
df['hour'] = df['transaction_time'].dt.strftime('%H:%M')

df = df.drop(columns=['transaction_time'])

print("Transformación exitosa ✅")
```

EDA (Ananlisis Exploratorio de datos)

```python
#Pregunta de negocio 1, cual es el medio por el cual hay mayor venta
ventas_por_canal = df['channel'].value_counts().reset_index()
ventas_por_canal.columns = ['channel', 'num_transacciones']
print(ventas_por_canal)
```

```python
#Horario de mayor venta
ventas_por_hora = df['hour'].value_counts().sort_index().reset_index()
ventas_por_hora.columns = ['hour', 'num_transacciones']
print(ventas_por_hora)
```

```python
#Categoria mas vendida
categoria_transacciones = (
    df['merchant_category']
    .value_counts()
    .reset_index()
)
categoria_transacciones.columns = ['merchant_category', 'num_transacciones']
print(categoria_transacciones.head(10))
```

A través del analisis exploratorio se obtiene la respuesta que el medio por cual mas se obtienen ventas es por la web en el horario de las 00:00 hs en la categoría de productos electronicos. Por lo cual se propone oferta para estos productos en dicho horaio y por dicho medio.
