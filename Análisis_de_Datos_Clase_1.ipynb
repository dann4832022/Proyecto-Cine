# Analisis de datos

## 1 ETL ( Extracción, Transformación y carga)

```python
from google.colab import drive
#drive.mount('/content/drive/drive/MyDrive', descarga (1)=True)

import pandas as pd
import chardet
ruta_archivo = '/content/drive/MyDrive/ventas.csv'
with open(ruta_archivo, 'rb') as f:
    result = chardet.detect(f.read())
    print(result)
try:
  df= pd.read_csv(ruta_archivo, encoding=result['encoding'], sep=';', decimal=',')
  print("Extraccion de datos exitoosa")
except FileExistsError:
  print("Error en la extraccion de datos")
```

```python
#Renombrar columnas para faciliar el acceso y evitar caracteres especiales
df.rename(columns={
    'Tienda': 'tienda',
    'Marca': 'marca',
    'Tipo': 'tipo',
    'Gï¿½nero': 'genero',
    'Talla': 'talla',
    'Color': 'color',
    'Categoria': 'categoria',
    'precio de venta': 'precio',
    'Fecha': 'fecha',
    'Hora': 'hora'
}, inplace=True)
```

```python
#convertir la columna precio a numérico
df['precio']=df['precio'].astype(str)
df['precio']=df['precio'].str.replace(',','.', regex=True).astype(float)


#print(set(df['genero'])) ver y obtener informacion resultado {'Usinex', 'Masculino', 'Femenino'}
#Cambiar el nombre de las columnas
df['genero'] = df['genero'].str.replace('Masculino','M', regex=False)
df['genero'] = df['genero'].str.replace('Femenino','F', regex=False)
df['genero'] = df['genero'].str.replace('Usinex','U', regex=False)

#Crear columnas de fecha y horas combinadas y convertilas al tipo datatime
df['fecha_hora'] = pd.to_datetime(df['fecha'] + ' ' + df['hora'], format='%d/%m/%Y %H:%M:%S')

print(df['talla'].unique())

#print(df.head())
#eliminar columnas de fecha y hora
df.drop(['fecha', 'hora'], axis=1, inplace=True)
#print(df.head())

#Dedocraticamenete tomamos las primera aparición numérica y descartamos el resto.
#Tomamos el valor de la izquerda si hay un guion

#df['talla'] = df['talla'].astype(str).apply(lambda x: x.split('-')[0])
#ahora lo convertimos a tipo numerico
#df['talla'] = pd.to_numeric(df['talla'], errors='coerce').astype('Int64')
#Eliminar fila con tallas mala  o no numéricas despues de la conversión
#df.dropna(subset=['talla'], inplace=True)

print(set(df['genero']))
print("Transformacion de datos exitosa")
```
